BETTER REPRESENTATIONS WITH DENSE VECTOR EMBEDDINGS

bag of words is a good method but it considers language to be nothing but a bag of words only, it doesnt capture the semantic nature of the text like what the text means..

word2vec released in 2013 was the first model to capture to meaning of texts in embeddings which are the vector rep of the data that tries to capture the meaning of the data, to  do this word2vec was trained on the vast amounts of textual information like the entire wikipedia

to make these sementic rep, word2vec uses neural networks see fig 1-6, using the several interconnected layers to process information, it generates word embeddings by seeing which words tends to come in next after a particular sentence

it starts by giving each word in the vocabulary a vector embedding, say 60 values assigned randomly to 60 words
then pairs are picked from the training data and the model predicts whether theres a chance they wil appear together in a sentence, see fig 1-7

during the training the model learns the relationship between the words and embedd them into the embedding,, if two words are tend to have same neighbours they will have closer embeddings and vice versa

when we say the embeddings tries to capture the meaning of the words we mean embeddings try to capture the information by representing the properties of words eg :
lets say we have two words "baby" and "apple", now the baby word will score higher in properties like "newborn" and "human", and apple will score low in these properties.

see fig 1-8 embeddings can have may properties at once as their sizes are fixed and properties help to create a mental representation of the word, it makes sense to a computer and an efficient way to understand and translate the human language

through embeddings we can also mesaure the semantic similarity between two words by using various distance based approach to see how close the words are
see fig 1-9 it shows the embedding in a 2d space as we can see the words with similar meaning tend to be closer

encoding and decoding context with attention

the word2vec creates a static representation of words, eg: the word "bank" will have the same embedding representation no matter the context in which it is used, as it can be used in financial context and also as a bank of river, therefore there was a need that the embeddings should change depending on the context

this type of text was formed with RNNs, it is a variant of neural networks which can model sequences to save the loss of context, meaning it saves the sequential context of the sentence, for eg :

the -> bank -> is -> that-> way

the sequence of the words caary a context, if we reshuffle the words the context will be lost
so RNNs takes all the words one by one produces the output and then inject that output in the leftover input to produce another output and the process goes on till we get all the outputs

now lets see it in our case, RNNs use two tasks encoding to represent an input sentence and decoding to output the ouput sentence see fig 1-11 which shows the translation of the sentence
each step in this arch is auto regressive like we talked previously, when generating the next word the model needs to get all previously generated words, see fig 1-12

the encoding step tries its best to represent the input as well as possible by generating the context in the form of an embedding,
now to generate a rep it takes the representations as input for decoder, so for the initial representations of the words we can use the word2vec model
see fig 1-13 we can see hoe the inputs are generated sequentially one at a time as well as the output
using word2vec embeddings a context embedding is generated that represent the entire sequence

now the earlier architecture used to compress the whole context in one embedding and pass it down to the decoder, eg workflow:

suppose we have a sentence to translate, the word2vec will create embeddings for each words in the sentence and pass it down to encoder

Attention is all you need

attention is the reason why todays LLMs have insane abilities, the authors of the paper proposed a arch called transformer which is base on the attention type network, it replaces the previous recurrent networks and also be trained parallely which significantly speeds up the training

the encoding and decodong components stacks up on top of each other see fig 1-16, this arch remains autoregressive which means it needs to consume each generated word before creating a new word

both the encoder and decoder uses attention instead of an RNN with attention features, the encoder part contains of two parts, self attention and a feed forward neural network shown in fig 1-17

now in attention RNNs the decoder attends to every encoder states before output genreation
but in transformer attention arch each input word attends to every other input words before any output generation

the self attention layer:
instead of processing one word at a time like RNN the transformer attends to all the words at once, so for each word the model determines the importance score for other words, the importance here means how much the other words are important to understand that particular word
for eg :

For "love" The model might determine:
Word	Importance
I	    30%
love	40%
llamas	30%
So the representation of "love" becomes a mixture of information from all three words.

For "llamas" The model might determine:
Word	Importance
I	    10%
love	50%
llamas	40%
Now "llamas" knows: Someone loves me.

For "I" The model might determine:
Word	Importance
I	    50%
love	40%
llamas	10%
Now "I" knows: I am the subject of loving.

after this interaction the embeddings become richer, and we get contextual representations shown as blue vectors in fig 1-17


it is called self attention coz the sentence attends to itself, thw words are simply exchanging information with each other no decoder or translation yet

self attention attends all the parts of the sentence so it can look in forward as well as backward direction unlike RNN which can look only at the backward direction

this way all are processed simultaneouly, the attention matrix is computed in parallel, this makes training faster

now in decoder also we have a self attention layer that attends the output of the encoder, see fig 1-19, this one is similar to the RNN attention decoder

fig 1-20 shows the decoder self attention layer masks the future positions so that there is no leakae of information while generating output

these building blocks create the transformer architecture and serves as the fondation for various impactful model in language AI such as BERT and GPT-1

REPRESENTATION MODELS :

1. ENCODER ONLY MODELS

the transormer is an encoder decoder architeture which is good for translation type task but cannot be easily implemented for other tasks like text classification

a new architecture was developed called bidirectional encoder representations for transformers (BERT) which will serve as a foundation for language AI,
it is an encoder only arch that focuses on representing language as shown in fig 1-21, there is no decoder involved

the encoder blocks are same as previous , self attention followed by feedforward neural networks, the input contains a token called [CLS] or classification token that represents the entire input, we use this token as input embedding for fine tuning the model on specific tasks like classification

train these encoders are difficult, BERT uses masked language modelling for this purpose, the fig 1-22 shows this process, some part of the input is masked and then model makes a prediction which is a difficult task but allows BERT to create more accurate intermediate rep of the input

this arch and the unique training process makes the bert and related architectures more suitable for contextual language representation, these models are used for transfer learning type tasks where we pretrain for language modelling and then fine tune it for specific task,
for eg : we train the model on the wikipedia to understand the contextual and semantic nature of the text.
fig 1-23 shows how can we then use the trained model for text classification task by fine tuning it

pretrained models does not need to train much as it is already trained, only fine tuning is required and training can be done using much less data
BERT like models generate embeddings at almost every step in the arch so they also act like feature extractor machines without the need of fine tuning them on specific tasks

they are also referred as representation models to diff them from decoder only model which are called generative models,
the main diff does not lies in arch or working, rep models create embeddings for the text and do not generate text like generative models

TRAINING PARADIGM OF LLMs

in traditional machine learning we train the model to perform one task alone like classification shown in fig 1-29, this is a one step process
but when it comes to LLM's there are two steps involved:

1. language modelling:
the first step is pretraining, the model is trained using a large collection of internet text which allows it to learn the correlation beteen words, the grammar, language patterns and the context, the training phase is not yet directed towards a specific task beyond predicting the next word, the trained model is called foundational model or base model which do not follow instructions

2. fine tuning
here the pretrained model is further trained for a specific task, this way the LLM can now focus on a particular task or show a desired behaviour
eg: we can fine tune a foundational model to perform a classification task or to follow instructions
this way we can save massive resources by skipping the train from scratch process completely again and again, good for people who cannot afford the compute, eg Llama 2 was trained on dataset of 2 trillion tokens just imagine the compute it wwould require to do that fig 1-30 shows the multistep approach in detail

lets use a generative model now, Phi-3-mini which is a relatively small model with 3.8 billion params, but its a good performing model

when we use an llm two models are loaded, the generative model itself and its underlying tokenizer

the tokenizer works by splitting the input text into tokens, before we give it to the generative model

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
"microsoft/Phi-3-mini-4k-instruct",
device_map="cuda",
torch_dtype="auto",
trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

we can use the pipeline to combine model, tokenizer and text generation into a single function

```python
from transformers import pipeline

# Create a pipeline
generator = pipeline(
"text-generation",
model=model,
tokenizer=tokenizer,
return_full_text=False,
max_new_tokens=500,
do_sample=False
)
```

return_full_text = false, this ensures only the output of the model returns without prompt

max_new tokens, the maximum no of tokens a model can generate, this way we can ensure the model dont produce long heavy outputs, as some models will coontinue generating output until it reaches its context window

do_sample, this tells whether the model will use any sampling strategy or not to select the next token, if we set it to false the model will always select the most probable token

to generate first text we will tell the model to joke about leftists, to do this we will prompt it using a list of dictionaries where each dic will  relate to an entity in conversation,

```python
# The prompt (user input / query)
messages = [
{"role": "user", "content": "Create a funny joke about leftists."}
]

# Generate output
output = generator(messages)
print(output[0]["generated_text"])
```
